# 網路使用經驗對成績的影響

## 將w1~w5, acheive拉出至新的df

```python
net_usage_df = df[['w1', 'w2', 'w3', 'w4', 'w5', 'achieve']].copy()
print("CheckIfNull:")
print(net_usage_df[['w1', 'w2', 'w3', 'w4', 'w5', 'achieve']].isnull().sum())
print("\nDescribe:")
print(net_usage_df[['w1', 'w2', 'w3', 'w4', 'w5', 'achieve']].describe())
```

```python
# 假設net_usage_df已經是讀入和準備好的DataFrame
# 變數名稱和描述
variables = {
    'w1': '沒有上網時心裡不舒服',
    'w2': '網路斷線時坐立不安',
    'w3': '上網時感到精神充沛',
    'w4': '沒有上網時情緒低落',
    'w5': '沒有上網時感到錯過'
}

# 創建一個圖形和多個子圖
fig, axes = plt.subplots(3, 2, figsize=(12, 18))  # 設定圖形大小和子圖的排列
axes = axes.flatten()  # 將子圖陣列扁平化，便於迭代

# 繪製每個變數的熱點圖
for i, (var, desc) in enumerate(variables.items()):
    # 計算交叉頻率表
    cross_tab = pd.crosstab(net_usage_df[var], net_usage_df['achieve'], normalize='index')

    # 繪製熱點圖
    sns.heatmap(cross_tab, ax=axes[i], cmap='viridis', annot=True)
    axes[i].set_title(f'{desc} vs. 學習成績表現')
    axes[i].set_xlabel('學習成績表現')
    axes[i].set_ylabel(desc)

# 移除多餘的子圖
fig.delaxes(axes[-1])

# 顯示圖形
plt.tight_layout()
plt.show()
```

根據這些熱點圖，我們可以觀察網路使用經驗變數（`w1` 到 `w5`）與學業表現（`achieve`）之間的關係。以下是對每個圖表的詳細分析和總體見解：

### 沒有上網時心裡不舒服 (w1)
此圖顯示了「沒有上網時心裡不舒服」對學業表現的影響。從熱點圖中可以看到，隨著「不舒服」程度的增加（從1到4），學業成績的分佈逐漸偏向中間區域，但並未呈現明顯的高低成績的集中現象。這意味著這種不適感可能對學業表現有一定的影響，但並不是主要因素。

### 網路斷線時坐立不安 (w2)
此圖展示了「網路斷線時坐立不安」對學業表現的影響。從圖中可以看到，隨著坐立不安程度的增加，成績在中等區間（6-7分）有較高的集中度，表明有較多學生在面臨網路斷線的焦慮時，仍能維持相對較高的學業表現。這可能反映出這些學生能夠有效地應對這種焦慮，並未大幅影響學業成績。

### 上網時感到精神充沛 (w3)
此圖展示了「上網時感到精神充沛」對學業表現的影響。從圖中可以看出，隨著感到精神充沛的程度增加（從1到4），學業成績分佈顯示出一定的正相關關係。這表明，對一些學生而言，上網能夠提供一定的精神振奮，從而對學業表現產生積極影響。

### 沒有上網時情緒低落 (w4)
此圖顯示了「沒有上網時情緒低落」對學業表現的影響。可以觀察到，隨著情緒低落程度的增加（從1到4），學業成績的高低變化並不顯著，這可能意味著情緒低落對學業表現的影響是間接的，並且可能與其他因素（如學習習慣、社交支持等）共同作用。

### 沒有上網時感到錯過 (w5)
此圖展示了「沒有上網時感到錯過」對學業表現的影響。從圖中可以看出，隨著感到錯過的程度增加（從1到4），學業成績在高分區域（7-8分）顯示出一定的集中度，這意味著感到錯過可能對部分學生產生動力作用，促使其更努力學習以彌補感知的損失。

### 總體見解
從這些圖表可以得出幾點結論：

1. **網路依賴與學業表現**：雖然一些網路使用經驗變數（如 `w1` 和 `w4`）顯示出對學業表現的潛在影響，但這些影響往往不是單一因素決定的。學生的學業表現受多種因素共同影響，網路使用經驗只是其中之一。

2. **正向和負向影響**：對於某些學生來說，上網可以提供精神上的振奮和支持（如 `w3`），這對學業表現有積極影響。而對於其他學生，過度依賴網路可能會引發焦慮和情緒低落，進而對學業表現產生負面影響。

3. **個體差異**：學生在面對網路使用經驗時的反應和其學業表現之間的關係存在明顯的個體差異。這表明在教育和心理輔導中，需要針對個體的具體情況進行個性化的支持和干預。



# 兩性互動經驗和色情網站的關聯性

```python
variables = ['learn13', 'opposit1', 'opposit2', 'opposit3', 'opposit4', 'netsex']
gender_df = df[variables].copy()
print("CheckIfNull:")
print(gender_df[variables].isnull().sum())
print("\nDescribe:")
print(gender_df[variables].describe())
```

```python
# 計算相關矩陣
corr_matrix = gender_df[variables].corr()
print(corr_matrix)
```

```python
# 繪製相關矩陣的熱點圖
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, cbar_kws={'label': 'Correlation Coefficient'})
plt.title('社交互動與網路使用之間的相關性熱點圖')
plt.xlabel('Variables')
plt.ylabel('Variables')

# 設定x軸和y軸的標籤
labels = ['與異性相處時害羞或不自在 (learn13)',
          '不知道如何與異性相處 (opposite1)',
          '認為男女社交機會太少 (opposite2)',
          '不知道如何處理異性交往中的偶發事件 (opposite3)',
          '在求學期間不想談戀愛但擔心畢業後沒機會 (opposite4)',
          '每週瀏覽色情網站的時間 (netsex)']
ax = plt.gca()
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.set_yticklabels(labels, rotation=0)

plt.tight_layout()
plt.show()
```

```python
import statsmodels.api as sm

# 準備數據
X = gender_df[['learn13', 'opposit1', 'opposit2', 'opposit3', 'opposit4']]
y = gender_df['netsex']

# 添加截距項
X = sm.add_constant(X)

# 建立回歸模型
model = sm.OLS(y, X).fit()

# 輸出回歸模型結果
print(model.summary())
```
